```python
import numpy as np
import pandas as pd
import math
```

```python
def gradient_descent(x,y):
    m_cur,b_cur = 0,0
    iter=15000
    n=len(x)
    ini_cost=float('inf')
    learn_rate =  0.0003  # 0.01 0.001 0.005 0.0001 trail and error
    for i in range(iter):
        y_pred = m_cur * x + b_cur
        cost = (1/n) *sum([val**2 for val in (y-y_pred)])
        if math.isclose(cost,ini_cost,rel_tol=1e-20, abs_tol=0.0):
            break
        ini_cost = cost
        m_der = -(2/n)*sum(x*(y-y_pred))  #derivative of m
        b_der = -(2/n)*sum(y-y_pred)       #derivate of b
        m_cur = m_cur - (learn_rate*m_der)   
        b_cur = b_cur - (learn_rate * b_der)
        print(m_cur,b_cur,cost,i)
```
